# Plotly Express

Plotly is a company that builds interactive and high-quality data visualization tools using Python, R, and JavaScript. 

It offers a huge variety of products, in particular the Dash, a framework for building web-based analytic applications. 

In 2019, they released a new high-level Python visualization library: Plotly Express.

The installation might be a little tricky, please follow the guide for Jupyter Lab [here](https://plotly.com/python/getting-started/).

The documentation can be accessed [here](https://plotly.com/python/plotly-express/).

## Importing and loading a sample dataset

```python
import plotly.express as px
import plotly.graph_objects as go

df = px.data.iris()
df.head()
```

## Simple scatter plot

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length")
fig.show()
```

## Scatter plot with colored categories and scatter size

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                     size='petal_length', hover_data=['petal_width'])
fig.show()
```

## 3D plots

```python
fig = px.scatter_3d(df, x="sepal_length", y="petal_length", z="petal_width", 
                    color="species",size='sepal_width')
fig.show()
```

## Relationship plots

```python
fig = px.scatter_matrix(df, dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], color="species")
fig.show()
```

## Simple boxplot

```python
fig = px.box(df, y="sepal_width")
fig.show()
```

## Category boxplot

```python
fig = px.box(df, x="species", y="sepal_width",color="species")
fig.show()
```

## Simple histogram

```python
fig = px.histogram(df, x="sepal_width")
fig.show()
```

## Category histogram

```python
fig = px.histogram(df, x="sepal_width", color="species")
fig.update_layout(barmode='overlay')
fig.update_traces(opacity=0.75)
fig.show()
```

## Combined scatter and boxplot

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", marginal_y="box",
           marginal_x="box", trendline="ols", template="simple_white")
fig.show()
```

## Combined scatter and histogram

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", marginal_y="histogram",
           marginal_x="histogram", trendline="ols", template="simple_white")
fig.show()
```

## Loading second dataset

```python
df_2 = px.data.gapminder()
df_2.head()
```

## Animated plots

```python
px.scatter(df_2, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country", 
           log_x = True, 
           size_max=45, range_x=[100,100000], range_y=[25,90])
```

## Define Plotly as default library when using Pandas

```python
import pandas as pd
pd.options.plotting.backend = "plotly"

df_2.plot(kind='bar', x='continent', y='pop')
```
